# **📊 2: Armor types distribution**

```python
%sql
SELECT 
    armor_type,
    COUNT(*) AS piece_count,
    ROUND(AVG(weight), 2) AS avg_weight
FROM eldenringcatalog.gold.dim_armors
GROUP BY armor_type
ORDER BY piece_count DESC;
```

**The armor weights follow a clear pattern based on coverage. Chest armor is the heaviest, averaging 8.91, making it the main source of protection. Leg armor comes next with an average weight of 5.74, followed by helms at 4.3. Gauntlets are the lightest at 3.25, reflecting their smaller size and limited coverage. Overall, weight decreases consistently from larger armor pieces to smaller ones.**

# **📊 3: Load armor stats and analyze physical negation**

```python
%sql
SELECT 
    armor_name,
    armor_type,
    negation_physical,
    negation_strike,
    negation_slash,
    negation_pierce,
    weight
FROM eldenringcatalog.gold.fact_armor_stats
ORDER BY negation_physical DESC
LIMIT 30;
```

**The top 30 armor pieces ranked by physical negation are dominated entirely by heavy chest armor, with values ranging from 20.4 down to 15.3. The Bull-Goat Armor and Verdigris Armor lead with the highest physical negation at 20.4, followed closely by other extremely heavy sets like Fire Prelate, Tree Sentinel, Banished Knight, and Radahn's Lion variants. Across all entries, high physical defense consistently pairs with strong strike, slash, and pierce negation, confirming that the tank-oriented, ultra-heavy chest pieces offer the best overall physical protection in the dataset.**

# **📊 4: Best physical protection per weight ratio**

```python
%sql
SELECT 
    armor_name,
    armor_type,
    negation_physical,
    weight,
    ROUND(negation_physical / weight, 2) AS protection_per_weight
FROM eldenringcatalog.gold.fact_armor_stats
ORDER BY protection_per_weight DESC
LIMIT 30;
```

**The best protection-per-weight ratios come almost entirely from lightweight gear, with the Circlet of Light outperforming all armor pieces at an exceptional 2.8 protection per weight. Very light chest pieces like the Ascetic’s Loincloth, Braided Cord Robe, and Goldmask’s Rags also rank highly, offering efficient physical defense despite their low overall stats. Several light leg pieces and helms appear as well, all staying under 3–5 weight. As the list progresses, slightly heavier but still efficient altered chest armors—such as Malenia’s Armor (Altered), Fire Knight Armor (Altered), and Raging Wolf Armor (Altered)—enter with solid ratios around 1.4. Overall, the data clearly shows that light armor provides the best physical protection relative to its weight, making it ideal for mobility-focused builds.**

# **📊 6: Poise per weight ratio**

```python
%sql
SELECT 
    armor_name,
    armor_type,
    resistance_poise,
    weight,
    ROUND(resistance_poise / weight, 2) AS poise_per_weight
FROM eldenringcatalog.gold.fact_armor_stats
ORDER BY poise_per_weight DESC
LIMIT 30;
```

**Poise efficiency is led by extremely lightweight pieces, with the Circlet of Light standing out at a massive 5 poise per weight, far higher than any other armor. Other top performers include the Ascetic’s Ankle Guards and Ascetic’s Loincloth, both offering unusually high poise for their very low weight. Beyond these outliers, many mid-weight chest armors—such as Fingerprint Armor, Mausoleum Knight Armor (Altered), and various Knight or Surcoat variants—deliver strong poise-per-weight ratios around 2.2–2.4, balancing sturdiness with manageable weight. Overall, the data shows that while heavy armor provides high total poise, the most weight-efficient poise comes from specific light or altered pieces, making them excellent choices for players who want poise without committing to full heavy load.**

# **📊 8: Best magic negation armors**

```python
%sql
SELECT 
    a.armor_type,
    a.armor_name,
    a.negation_magic,
    a.weight
FROM eldenringcatalog.gold.fact_armor_stats a
INNER JOIN (
    SELECT armor_type, MAX(negation_magic) as max_magic
    FROM eldenringcatalog.gold.fact_armor_stats
    GROUP BY armor_type
) max_vals ON a.armor_type = max_vals.armor_type AND a.negation_magic = max_vals.max_magic
ORDER BY a.negation_magic DESC;
```

**Across all armor categories, the highest magic negation pieces are dominated by lightweight caster gear. Azur's Glintstone Robe and Lusat's Robe tie for the strongest overall protection with 15.4 magic negation at just 7.1 weight, making chest robes the best option for maximizing magical defense. For leg armor, Preceptor's Trousers lead with 8.3 magic negation at 3.9 weight, while the Nox Mirrorhelm provides the best magic negation among helms with 6.7, though at a slightly heavier 7.5 weight. Finally, Preceptor’s Gloves stand out among gauntlets with 3.6 magic negation at only 2.1 weight. Overall, the best magic defense per armor type consistently comes from light, mage-oriented gear, offering strong protection without the burden of heavy load.**

```python
%sql
SELECT armor_name, armor_type,
    (negation_physical + negation_strike + negation_slash + negation_pierce + negation_magic + negation_fire + negation_lightning + negation_holy) AS total_negation,
    (resistance_immunity + resistance_robustness + resistance_focus + resistance_vitality + resistance_poise) AS total_resistance,
    (negation_physical + negation_strike + negation_slash + negation_pierce + negation_magic + negation_fire + negation_lightning + negation_holy) + 
    ((resistance_immunity + resistance_robustness + resistance_focus + resistance_vitality + resistance_poise) * 0.5) AS defense_score,
    weight
FROM eldenringcatalog.gold.fact_armor_stats
ORDER BY defense_score DESC LIMIT 30;
```

**The highest overall defense scores are dominated by heavy chest armors, with Fire Prelate Armor, Armor of Solitude, Verdigris Armor, and Bull-Goat Armor leading the list. These armors combine very high physical and elemental negation with strong overall resistances, giving them the top defense scores of around 280–290, though they come with substantial weight (20–26.5). Mid-tier chest armors like Omen Armor, Gaius’s Armor, and Lionel’s Armor also provide excellent all-around protection, balancing solid negation and resistance with slightly lower weight. Lightweight magical or specialty robes, such as Lusat’s Robe, Azur’s Glintstone Robe, and Mushroom Body, achieve respectable defense scores (~217) despite their low weight, highlighting their efficiency for mobility-focused builds. Overall, heavy chest armor maximizes total defense, while light mage-focused gear offers a highly weight-efficient alternative with decent protection.**

```python
%sql
SELECT 
    CASE WHEN a.is_dlc = 1 THEN 'DLC' ELSE 'Base Game' END AS content_type,
    COUNT(*) AS armor_count,
    ROUND(AVG(s.weight), 2) AS avg_weight,
    ROUND(AVG(s.negation_physical), 2) AS avg_physical,
    ROUND(AVG(s.resistance_poise), 2) AS avg_poise
FROM eldenringcatalog.gold.dim_armors a
INNER JOIN eldenringcatalog.gold.fact_armor_stats s ON a.armor_name = s.armor_name
GROUP BY a.is_dlc;
```

**Across Elden Ring, DLC and base game armors are very similar on average. DLC armors (145 pieces) have an average weight of 6.02, physical negation of 6.09, and poise of 10.82, while base game armors (578 pieces) average 5.99 weight, 6.12 physical negation, and 10.56 poise. This indicates that DLC armors maintain comparable defensive performance to base game gear, with only a slight increase in poise on average.**

```python
%sql
SELECT armor_name, armor_type, weight,
       (negation_physical + negation_strike + negation_slash + negation_pierce +
        negation_magic + negation_fire + negation_lightning + negation_holy) AS total_negation,
       (resistance_immunity + resistance_robustness + resistance_focus + resistance_vitality + resistance_poise) AS total_resistance,
       ((negation_physical + negation_strike + negation_slash + negation_pierce +
         negation_magic + negation_fire + negation_lightning + negation_holy) + 
         (resistance_immunity + resistance_robustness + resistance_focus + resistance_vitality + resistance_poise)*0.5) AS defense_score
FROM eldenringcatalog.gold.fact_armor_stats
WHERE weight BETWEEN 10 AND 25
ORDER BY defense_score DESC
LIMIT 20;
```

**the balanced armor sets with moderate weight between 10 and 25 units, prioritizing overall defense efficiency. The top entries are predominantly chest armor pieces, with Fire Prelate Armor leading with a defense score of 289.9, followed closely by Armor of Solitude (285.8) and Omen Armor (279.9). These armors combine high total negation against physical, elemental, and magical damage with substantial resistances in immunity, robustness, focus, vitality, and poise. Other notable armors include Gaius’s Armor, Lionel’s Armor, and Tree Sentinel Armor, all offering strong protection without excessively burdening the wearer. Overall, this selection demonstrates a careful balance between weight and defensive capabilities, making them ideal for players seeking robust protection while maintaining moderate mobility.**
