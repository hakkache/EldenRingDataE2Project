# **📊 3: Analyze FP cost distribution**


**✔️ Simple Explanation**

**If a spell has:**

**fp_cost_min → the minimum FP it uses (usually a tap or quick cast)**

**fp_cost_max → the maximum FP it uses (usually a charged, held, or extended cast)**

```python
%sql
SELECT 
    DISTINCT spell_name,
    spell_type,
    fp_cost_min,
    fp_cost_max,
    fp_cost_per_use
FROM eldenringcatalog.gold.fact_spell_usage
ORDER BY fp_cost_min DESC
LIMIT 30;
```

```python
%sql
SELECT 
    AVG(fp_cost_min) AS avg_fp,
    MIN(fp_cost_min) AS min_fp,
    MAX(fp_cost_min) AS max_fp,
    STDDEV(fp_cost_min) AS stddev_fp
FROM eldenringcatalog.gold.fact_spell_usage;
```

**The analysis of spell FP consumption shows that the most demanding abilities are primarily high-tier incantations, with Erdtree Heal topping the list at 65 FP, followed by powerful boss-tier spells like Placidusax’s Ruin and Blessing of the Erdtree. Legendary sorceries such as Ranni’s Dark Moon and Rellana’s Twin Moons also appear among the highest-cost spells. Notably, several dragon incantations (Ekzykes’s Decay, Borealis’s Mist, Rotten Breath, Agheel’s Flame) have variable FP usage, allowing players to extend casting duration. When looking at overall spell data, spells average 23.67 FP, with a minimum of 1 FP, a maximum of 65 FP, and a standard deviation of 12.44, indicating a wide spread between basic utility spells and high-end, late-game abilities. Overall, FP consumption increases dramatically for advanced healing, buffs, dragon breaths, and legendary sorceries.**

# **📊 10: Spells requiring multiple stats**

```python
%sql
SELECT 
    distinct spell_name,
    spell_type,
    required_intelligence,
    required_faith,
    required_arcane,
    (CASE WHEN required_intelligence > 0 THEN 1 ELSE 0 END +
     CASE WHEN required_faith > 0 THEN 1 ELSE 0 END +
     CASE WHEN required_arcane > 0 THEN 1 ELSE 0 END) AS stat_count
FROM eldenringcatalog.gold.fact_spell_usage
WHERE (CASE WHEN required_intelligence > 0 THEN 1 ELSE 0 END +
       CASE WHEN required_faith > 0 THEN 1 ELSE 0 END +
       CASE WHEN required_arcane > 0 THEN 1 ELSE 0 END) >= 2
ORDER BY stat_count DESC;
```

**Many Elden Ring spells rely on multiple stats rather than a single attribute. These dual-stat spells, requiring combinations of Intelligence, Faith, and Arcane, provide versatile effects. For example, Radagon's Rings of Light uses both Intelligence and Faith, while Arcane-heavy Incantations like Glintstone Breath pair Arcane with Faith. Sorceries such as Explosive Ghostflame combine Intelligence and Faith for powerful damage. Investing in two stats can unlock these potent hybrid spells, offering flexibility and stronger scaling for builds that aren’t purely single-stat focused.**

# **📊 11: Calculate total stat requirements per spell**

```python
%sql
SELECT 
    distinct spell_name,
    spell_type,
    required_intelligence,
    required_faith,
    required_arcane,
    (required_intelligence + required_faith + required_arcane) AS total_stat_requirement,
    fp_cost_min
FROM eldenringcatalog.gold.fact_spell_usage
ORDER BY total_stat_requirement DESC
LIMIT 30;
```

**The spells with the highest total stat requirements in Elden Ring demand significant investment in one or more attributes. For instance, Light of Miquella and Explosive Ghostflame each require a combined 72 points across Intelligence and Faith, while classic Sorceries like Rellana's Twin Moons and Rennala's Full Moon demand high Intelligence alone. Hybrid spells like Radagon's Rings of Light use both Intelligence and Faith, offering flexibility but still requiring careful stat allocation. These high-requirement spells often have powerful effects or unique mechanics, making them highly rewarding for players willing to invest in the necessary stats.**

# **📊 12: FP efficiency metric (FP cost per stat requirement)**

```python
%sql
SELECT 
    distinct spell_name,
    spell_type,
    fp_cost_min,
    (required_intelligence + required_faith + required_arcane) AS total_reqs,
    ROUND(fp_cost_min / (required_intelligence + required_faith + required_arcane), 2) AS fp_per_stat
FROM eldenringcatalog.gold.fact_spell_usage
WHERE (required_intelligence + required_faith + required_arcane) > 0 
  AND fp_cost_min > 0
ORDER BY fp_per_stat
LIMIT 30;
```

**the most FP-efficient spells in Elden Ring, showing which spells consume the least FP relative to their total stat requirements. Spells like Surge, O Flame! and Discus of Light have extremely low FP-per-stat ratios (0.11–0.13), making them very cost-effective for players who want high utility without burning through FP. On the other hand, high-powered spells like Radagon's Rings of Light or Explosive Ghostflame still maintain reasonable FP efficiency (0.34–0.40) considering their high total stat requirements. This kind of analysis helps players optimize their builds, balancing spell power with sustainability in combat.**.

# **📊 24: Calculate "spell value" score**

```python
%sql
SELECT 
    distinct spell_name,
    spell_type,
    fp_cost_min,
    (required_intelligence + required_faith + required_arcane) AS total_reqs,
    slots_required,
    ROUND(100 / ((fp_cost_min * 0.5) + ((required_intelligence + required_faith + required_arcane) * 0.3) + (slots_required * 10)), 2) AS value_score
FROM eldenringcatalog.gold.fact_spell_usage
WHERE fp_cost_min > 0 
    AND (required_intelligence + required_faith + required_arcane) > 0
ORDER BY value_score DESC
LIMIT 30;
```

**The highest-ranked spell is “Surge, O Flame!” (Incantation) with a minimal FP cost of 1, low total stat requirement of 9, and just 1 slot, yielding a value score of 7.58. Other top contenders include Swift Glintstone Shard, Carian Slicer, Cure Poison, Glintstone Pebble, and Bestial Sling, all combining low FP, moderate stat requirements, and efficient slot usage.**

**These spells represent high-efficiency choices for builds that want maximum utility with minimal resource investment, perfect for early-to-mid game or hybrid casters seeking strong performance without overcommitting stats or FP**

# **📊 26: Spell recommendations by build**

## INTELLIGENCE BUILD (Best sorceries)

```python
%sql
SELECT 
    distinct spell_name,
    required_intelligence,
    fp_cost_min,
    bonus,
    ROUND(required_intelligence / (fp_cost_min + 1), 2) AS efficiency
FROM eldenringcatalog.gold.fact_spell_usage
WHERE spell_type = 'Sorcery' AND required_intelligence > 0
ORDER BY efficiency DESC
LIMIT 10;
```

## FAITH BUILD (Best incantations)

```python
%sql
SELECT 
    distinct spell_name,
    required_faith,
    fp_cost_min,
    bonus,
    ROUND(required_faith / (fp_cost_min + 1), 2) AS efficiency
FROM eldenringcatalog.gold.fact_spell_usage
WHERE spell_type = 'Incantation' AND required_faith > 0
ORDER BY efficiency DESC
LIMIT 10;
```

## Recommendations
### Intelligence (Pure Sorcery)
- Early core: Glintstone Pebble, Carian Slicer (cheap spam + melee reach).
- Mid-game: Rock Sling (gravity stagger), Swift Glintstone Shard (speed), Loretta’s spells (range).
- Late nukes: Comet Azur, Full Moon trilogy, Rellana’s Twin Moons (reserve 3-slot only if replacing other burst).
- Slot strategy: 1 high-impact (Full Moon / Comet Azur) + 2–3 low-FP sustain sorceries for dungeon clears.

### Faith (Pure Incantation)
- Early sustain: Catch Flame, Flame Sling, Heal / Urgent Heal, Cure Poison utility.
- Mid utility: Golden Vow, Blessing of the Erdtree (long buffs), Flame Grant Me Strength (hybrid AR boost).
- Late power: Light of Miquella, Minor Erdtree, Messmer’s Orb (situational boss usage).
- Keep one cheap offensive + one AoE + one buff; avoid stacking multiple long-cast heals in solo play.

### Arcane / Dragon / Blood
- Status application: Rotten Breath / Ekzykes’s Decay (rot), Swarm of Flies (bleed pressure).
- Debuff: Greyoll’s Roar (use sparingly—high stamina).
- High investment burst: Bayle’s Flame Lightning / Bayle’s Tyranny (consider flask allocation for FP sustain).
- Mix one rot/breath, one utility debuff, one direct damage lightning.

### Hybrid INT/FAI
- Versatile core: Radagon’s Rings of Light (fast holy beam), Triple Rings of Light (multi-lane control), Explosive Ghostflame (AoE DOT).
- Catalyst choice: Use Golden Order Seal or Prince of Death’s staff/seal pairing depending on scaling focus.
- Limit to 1–2 dual-stat high-cost spells; fill remaining slots with a cheap mono-stat tool for resource efficiency.

### Efficiency / Value Optimization
- Mob clear set: Surge, O Flame! (tick), Carian Slicer (melee), Swift Glintstone Shard / Glintstone Pebble (ranged).
- High value rotation: Alternate a low FP spell every other cast to preserve flasks in long dungeons.
- Track FP-per-stat; replace spells >0.6 ratio if a lower alternative covers similar role.

### Memory Slot Management
- Reserve 3-slot spells only if build centered on them (Comet Azur beam setups, Placidusax’s Ruin arena control).
- Typical optimal layout: 1 burst (large slot or high FP), 2–3 efficiency fillers, 1 buff, 1 utility/status.

### Stamina & Casting Flow
- Avoid chaining multiple high stamina spells before weapon attacks; stagger Greyoll’s Roar at opener then swap to low-stamina options.
- Use high stamina-to-FP ratio spells (Discus of Light, Briars of Sin) for sustained pressure in longer fights.

### Flask & Resource Strategy
- INT builds: Allocate slightly more FP flasks if running multiple high-cost nukes (e.g., Comet Azur + Full Moon).
- Faith builds: Balance HP vs FP; large heals reduce FP flask need—keep at least one low-FP emergency heal.
- Hybrid: Even flask split; rely on efficient value spells to prevent FP starvation.

### DLC Integration
- Faith builds gain stronger late-game scaling via DLC additions—justify pushing Faith >60 earlier.
- Reassess stat allocation: pure INT may add one hybrid beam (Radagon’s Rings) if leveling Faith marginally.

### Progression Path
1. Start: Secure cheap spam (Pebble / Catch Flame) + one utility (Heal / Cure Poison).
2. Mid: Introduce a control spell (Gravity / Rot) + buff (Golden Vow).
3. Late: Add signature nuke (Comet Azur / Light of Miquella) and optimize slot economy by pruning redundant mid-tier damage spells.

### Risk Mitigation
- High FP & multi-slot spells reduce flexibility—test encounter efficacy before permanent slot commitment.
- Preserve at least one rapid-cast, low commitment spell for interrupt windows and finishing blows.
