# **📊  4- What is the weight distribution of weapons?**

**Weight affects equipment load and mobility.**

```python
%sql
SELECT 
    CASE 
        WHEN weight <= 3 THEN 'Light'
        WHEN weight <= 8 THEN 'Medium'
        WHEN weight <= 15 THEN 'Heavy'
        ELSE 'Very Heavy'
    END AS weight_class,
    COUNT(*) AS weapon_count,
    ROUND(AVG(weight), 2) AS avg_weight
FROM eldenringcatalog.gold.dim_weapons
GROUP BY weight_class
ORDER BY avg_weight;
```

**Most weapons fall into medium weight range, with extremes for specialized builds.**

# **11: What is the highest base physical attack damage?**

**Raw damage output at standard upgrade level.**


 All normal weapons start as STANDARD

 Affinities like Sacred, Flame, Cold, Fire, Keen, etc. modify the base form

 Unique/Somber weapons do not have affinities

 If upgrade level has no "+" and affinity is Standard  its the true base weapon

```python
%sql
SELECT 
    fs.weapon_name,
    fs.category,
    w.is_dlc,
    fs.upgrade_level,
    fs.attack_physical,
    fs.weight,
    fs.total_attack_power
FROM eldenringcatalog.gold.fact_weapon_stats fs
JOIN eldenringcatalog.gold.dim_weapons w
ON fs.weapon_name = w.weapon_name
WHERE fs.upgrade_level LIKE '%Standard%'
AND fs.Upgrade_Level NOT LIKE '%+%'
ORDER BY fs.attack_physical DESC
```

**Colossal swords and other heavy weapons usually offer the highest physical damage, with the Jar Cannon having the highest base physical attack**

# **📊 12 - Calculate damage per weight ratio for all weapons**

**Efficiency metric for mobility-focused builds.**

```python
%sql
SELECT 
    weapon_name,
    category,
    attack_physical,
    weight,
    ROUND(
        CASE 
            WHEN weight = 0 THEN NULL
            ELSE attack_physical / weight
        END, 2
    ) AS damage_per_weight
FROM eldenringcatalog.gold.fact_weapon_stats
WHERE upgrade_level LIKE '%Standard%'
AND upgrade_level NOT LIKE '%+%'
ORDER BY damage_per_weight DESC
LIMIT 30;
```

```python
%sql
SELECT 
    category,count( *) as count,
    MAX(
        CASE 
            WHEN weight = 0 THEN NULL
            ELSE ROUND(attack_physical / weight, 2)
        END
    ) AS best_efficiency
FROM eldenringcatalog.gold.fact_weapon_stats
WHERE upgrade_level LIKE '%Standard%'
AND upgrade_level NOT LIKE '%+%'
GROUP BY category
ORDER BY best_efficiency DESC;
```

**Light weapons often have better damage-per-weight ratios, favoring agile builds**

# **📊 19 - Which weapon has the highest total attack power at max upgrade?**

```python
%sql
SELECT weapon_name,
       category,
       upgrade_level,
       total_attack_power,
       attack_physical,
       attack_magic,
       attack_fire,
       attack_lightning,
       attack_holy
FROM (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY weapon_name ORDER BY total_attack_power DESC) AS rn
    FROM eldenringcatalog.gold.fact_weapon_stats
) t
WHERE rn = 1
ORDER BY total_attack_power DESC
LIMIT 20;
```

# **📊 Question 25: Calculate average upgrade progression for physical damage**

**Understand damage scaling through upgrade levels.**

```python
%sql
-- 1) Extract numeric upgrade number for Standard rows
WITH base AS (
    SELECT
        weapon_name,
        upgrade_level,
        TRY_CAST(regexp_extract(upgrade_level, '\\+(\\d+)', 1) AS INT) AS upgrade_number,
        attack_physical
    FROM eldenringcatalog.gold.fact_weapon_stats
    WHERE upgrade_level LIKE 'Standard%'
      AND regexp_extract(upgrade_level, '\\+(\\d+)', 1) <> ''   -- only rows containing +N
),

-- 2) Keep rows with valid upgrade_number
clean AS (
    SELECT *
    FROM base
    WHERE upgrade_number IS NOT NULL
),

-- 3) Determine the max Standard upgrade number per weapon
weapon_max AS (
    SELECT
        weapon_name,
        MAX(upgrade_number) AS max_upgrade
    FROM clean
    GROUP BY weapon_name
),

-- 4) Assign each weapon to its upgrade path (Standard+10 or Standard+25)
with_path AS (
    SELECT
        c.weapon_name,
        c.upgrade_level,
        c.upgrade_number,
        c.attack_physical,
        wm.max_upgrade,
        CONCAT('Standard+', CAST(wm.max_upgrade AS STRING)) AS upgrade_path
    FROM clean c
    JOIN weapon_max wm
      ON c.weapon_name = wm.weapon_name
),

-- 5) Aggregate: average physical damage per upgrade_number per path
agg AS (
    SELECT
        upgrade_path,
        upgrade_number,
        ROUND(AVG(attack_physical), 2) AS avg_physical_damage,
        COUNT(*) AS weapon_count
    FROM with_path
    GROUP BY upgrade_path, upgrade_number
),

-- 6) Compute percent progression within each (Standard+X) path
progression AS (
    SELECT
        upgrade_path,
        upgrade_number,
        avg_physical_damage,
        LAG(avg_physical_damage) OVER (
            PARTITION BY upgrade_path ORDER BY upgrade_number
        ) AS prev_damage
    FROM agg
)

-- 7) Final output
SELECT
    upgrade_path,
    upgrade_number,
    avg_physical_damage,
    ROUND(
        CASE
            WHEN prev_damage IS NULL THEN 0
            ELSE ((avg_physical_damage - prev_damage) / prev_damage) * 100
        END,
        2
    ) AS pct_increase
FROM progression
ORDER BY upgrade_path, upgrade_number;
```

**See how damage scales with each upgrade level, plan upgrade priorities.**
