# **📊 12 - Calculate damage per weight ratio for all weapons**

**Efficiency metric for mobility-focused builds.(base without upgrade)**

```python
%sql
SELECT 
    weapon_name,
    category,
    attack_physical,
    weight,
    ROUND(
        CASE 
            WHEN weight = 0 THEN NULL
            ELSE attack_physical / weight
        END, 2
    ) AS damage_per_weight
FROM eldenringcatalog.gold.fact_weapon_stats
WHERE upgrade_level LIKE '%Standard%'
AND upgrade_level NOT LIKE '%+%'
ORDER BY damage_per_weight DESC
LIMIT 30;
```

```python
%sql
SELECT 
    category,count( *) as count,
    MAX(
        CASE 
            WHEN weight = 0 THEN NULL
            ELSE ROUND(attack_physical / weight, 2)
        END
    ) AS best_efficiency
FROM eldenringcatalog.gold.fact_weapon_stats
WHERE upgrade_level LIKE '%Standard%'
AND upgrade_level NOT LIKE '%+%'
GROUP BY category
ORDER BY best_efficiency DESC;
```

# **📊 Question 25: Calculate average upgrade progression for physical damage**

**Understand damage scaling through upgrade levels.**

```python
%sql
-- 1) Extract numeric upgrade number for Standard rows
WITH base AS (
    SELECT
        weapon_name,
        upgrade_level,
        TRY_CAST(regexp_extract(upgrade_level, '\\+(\\d+)', 1) AS INT) AS upgrade_number,
        attack_physical
    FROM eldenringcatalog.gold.fact_weapon_stats
    WHERE upgrade_level LIKE 'Standard%'
      AND regexp_extract(upgrade_level, '\\+(\\d+)', 1) <> ''   -- only rows containing +N
),

-- 2) Keep rows with valid upgrade_number
clean AS (
    SELECT *
    FROM base
    WHERE upgrade_number IS NOT NULL
),

-- 3) Determine the max Standard upgrade number per weapon
weapon_max AS (
    SELECT
        weapon_name,
        MAX(upgrade_number) AS max_upgrade
    FROM clean
    GROUP BY weapon_name
),

-- 4) Assign each weapon to its upgrade path (Standard+10 or Standard+25)
with_path AS (
    SELECT
        c.weapon_name,
        c.upgrade_level,
        c.upgrade_number,
        c.attack_physical,
        wm.max_upgrade,
        CONCAT('Standard+', CAST(wm.max_upgrade AS STRING)) AS upgrade_path
    FROM clean c
    JOIN weapon_max wm
      ON c.weapon_name = wm.weapon_name
),

-- 5) Aggregate: average physical damage per upgrade_number per path
agg AS (
    SELECT
        upgrade_path,
        upgrade_number,
        ROUND(AVG(attack_physical), 2) AS avg_physical_damage,
        COUNT(*) AS weapon_count
    FROM with_path
    GROUP BY upgrade_path, upgrade_number
),

-- 6) Compute percent progression within each (Standard+X) path
progression AS (
    SELECT
        upgrade_path,
        upgrade_number,
        avg_physical_damage,
        LAG(avg_physical_damage) OVER (
            PARTITION BY upgrade_path ORDER BY upgrade_number
        ) AS prev_damage
    FROM agg
)

-- 7) Final output
SELECT
    upgrade_path,
    upgrade_number,
    avg_physical_damage,
    ROUND(
        CASE
            WHEN prev_damage IS NULL THEN 0
            ELSE ((avg_physical_damage - prev_damage) / prev_damage) * 100
        END,
        2
    ) AS pct_increase
FROM progression
ORDER BY upgrade_path, upgrade_number;
```

**after examining physical damage progression for Standard weapons along their maximum upgrade paths, Standard+10 and Standard+25. For the Standard+10 path, weapons start at around 112 physical damage at +1 and reach approximately 243 at +10, with the largest increases occurring at lower upgrades—12.9% from +1 to +2—gradually tapering to about 6.4% by +10. In the Standard+25 path, weapons begin slightly lower at 109 but scale up to 255 by +25, with early upgrades providing moderate gains of 5–4% that steadily decline to 2.4% at the maximum level. Overall, the data shows that physical damage growth is strongest in the early upgrades and diminishes as weapons near their peak, while longer upgrade paths offer more incremental gains over a wider range.**

```python
%sql
WITH ranked AS (
    SELECT
        category,
        weapon_name,
        weight,
        ROW_NUMBER() OVER (
            PARTITION BY category
            ORDER BY weight
        ) AS rank
    FROM
        eldenringcatalog.gold.fact_weapon_stats
)
SELECT
    category,
    weapon_name,
    weight
FROM ranked
WHERE rank = 1
ORDER BY weight;
```

# **📊 28: Which weapon category requires the highest average stats?**

**Understanding accessibility barriers per weapon type.**

```python
%sql
SELECT
    category,
    ROUND(AVG(required_strength), 2)      AS avg_str,
    ROUND(AVG(required_dexterity), 2)     AS avg_dex,
    ROUND(AVG(required_intelligence), 2)  AS avg_int,
    ROUND(AVG(required_faith), 2)         AS avg_fai,
    ROUND(AVG(required_arcane), 2)        AS avg_arc,
    COUNT(*) AS weapon_count,
    (
        ROUND(AVG(required_strength), 2) +
        ROUND(AVG(required_dexterity), 2) +
        ROUND(AVG(required_intelligence), 2) +
        ROUND(AVG(required_faith), 2) +
        ROUND(AVG(required_arcane), 2)
    ) AS avg_total_reqs
FROM eldenringcatalog.gold.dim_weapons
GROUP BY category
ORDER BY avg_total_reqs DESC;
```

# **📊 Question 37: Average damage increase from base to +10 upgrade**

**ROI on upgrade materials.**

```python
%sql
WITH base AS (
    SELECT
        weapon_name,
        attack_physical AS base_damage
    FROM eldenringcatalog.gold.fact_weapon_stats
    WHERE upgrade_level LIKE '%Standard '
),
plus10 AS (
    SELECT
        weapon_name,
        attack_physical AS plus10_damage
    FROM eldenringcatalog.gold.fact_weapon_stats
    WHERE upgrade_level LIKE '%Standard +10%'
),
upgrade_roi AS (
    SELECT
        b.weapon_name,
        b.base_damage,
        p.plus10_damage,
        (p.plus10_damage - b.base_damage) AS damage_increase,
        ROUND(
            CASE 
                WHEN b.base_damage = 0 THEN 0
                ELSE ((p.plus10_damage - b.base_damage) / b.base_damage) * 100
            END,
            2
        ) AS pct_increase
    FROM base b
    INNER JOIN plus10 p
        ON b.weapon_name = p.weapon_name
)

SELECT *
FROM upgrade_roi
ORDER BY pct_increase DESC
LIMIT 30;

-- SELECT ROUND(AVG(pct_increase), 2) AS avg_pct_increase
-- FROM upgrade_roi;
```

**This query analyzes the damage increase from Standard to Standard +10 upgrades for weapons. It calculates both the absolute and percentage increase in physical attack.**

**Key takeaways from the top 30 weapons by upgrade ROI:**

**Some crossbows and ballistas show the highest percentage increase, around 250% (e.g., Crepus's Black-Key Crossbow, Full Moon Crossbow, Jar Cannon, Rabbath's Cannon), indicating huge scaling from upgrades.**

**Many melee weapons have a ~145% increase (e.g., Greatsword of Solitude, Obsidian Lamina, Dragon Greatclaw, Morgott's Cursed Sword), showing substantial gains from +10 upgrades.**

**Percentage increase is computed as (plus10_damage - base_damage) / base_damage * 100, highlighting weapons with the best return on investment when fully upgraded.**

**This helps identify weapons that benefit most from upgrading, both ranged and melee.**

**In short, crossbows, ballistas, and certain heavy melee weapons see the largest damage jumps from Standard → Standard +10, making them highly rewarding for full upgrade investment.**

```python
%sql
WITH weapon_reqs AS (
    SELECT
        weapon_name,
        category,
        required_strength,
        required_dexterity,
        required_intelligence,
        required_faith,
        required_arcane,
        (required_strength + required_dexterity + required_intelligence + required_faith + required_arcane) AS total_reqs,
        GREATEST(required_strength, required_dexterity, required_intelligence, required_faith, required_arcane) AS max_req,
        LEAST(required_strength, required_dexterity, required_intelligence, required_faith, required_arcane) AS min_req
    FROM eldenringcatalog.gold.dim_weapons
)
SELECT
    weapon_name,
    category,
    required_strength,
    required_dexterity,
    required_intelligence,
    required_faith,
    required_arcane,
    (max_req - min_req) AS req_range
FROM weapon_reqs
WHERE total_reqs > 20
ORDER BY req_range
LIMIT 20;
```

 **The query highlights weapons in Elden Ring whose total attribute requirements exceed 20 and calculates the difference between the weapon's highest and lowest required stat. The result lists the 20 weapons with the smallest to moderate differences in required stats, meaning these weapons demand a relatively balanced spread of attributes across Strength, Dexterity, Intelligence, Faith, and Arcane. For example, weapons like Smithscript Dagger and Frenzyflame Perfume Bottle have multiple moderate stat requirements, while others like Beast Claw (Weapon) or Wakizashi focus more on physical stats but still require some spread. This indicates these weapons are suitable for hybrid builds, where players need to invest in several stats rather than specializing in only one.**

# **📊 Question 49: Calculate "total offense" score (attack power + critical + buildup)**

**Comprehensive damage potential metric.**

```python
%sql
WITH max_buildup_per_weapon AS (
    SELECT
        weapon_name,
        MAX(total_buildup) AS max_buildup
    FROM eldenringcatalog.gold.fact_status_effects
    WHERE upgrade_level LIKE '%Standard%'
    AND upgrade_level NOT LIKE '%+%'
    GROUP BY weapon_name
)

SELECT
    ws.weapon_name,
    ws.category,
    ws.total_attack_power,
    ws.critical_damage,
    COALESCE(mb.max_buildup, 0) AS max_buildup,
    (ws.total_attack_power * 1.0) 
      + (ws.critical_damage * 0.5) 
      + (COALESCE(mb.max_buildup, 0) * 2.0)
      - (ws.stamina_cost * 0.5) AS offense_score
FROM eldenringcatalog.gold.fact_weapon_stats ws
LEFT JOIN max_buildup_per_weapon mb
    ON ws.weapon_name = mb.weapon_name
WHERE ws.upgrade_level LIKE '%Standard%'
AND upgrade_level NOT LIKE '%+%'
ORDER BY offense_score DESC
LIMIT 30;
```

**The query ranks base (non-upgraded) weapons in Elden Ring by an “offense score”, which combines total attack power, critical damage, status buildup, and stamina cost into a single metric. The top weapons include Halo Scythe, Winged Scythe, and Vyke's War Spear, each excelling due to a mix of high base attack and significant status buildup. Other strong weapons in this list are Rotten Staff, St. Trina's Torch, Nanaya's Torch, and Meteoric Ore Blade. The offense score formula values status buildup heavily, giving weapons that inflict bleed, poison, or other effects a higher score, while penalizing stamina cost slightly. This ranking is based solely on base, standard-level weapons, without considering any +1 or higher upgrades, showing which weapons are innately powerful before enhancements.**

# **📊 Question 50: Create weapon recommendation matrix by build type**

**Final comprehensive guide for weapon selection.**

```python
%sql
WITH df_recommendations AS (
    SELECT
        ws.weapon_name,
        ws.category,
        ws.scaling_strength,
        ws.scaling_dexterity,
        ws.scaling_intelligence,
        ws.scaling_faith,
        w.required_strength,
        w.required_dexterity,
        w.required_intelligence,
        w.required_faith,
        w.required_arcane,
        
        -- Pure STR score
        (CASE 
            WHEN ws.scaling_strength IN ('S', 'A') THEN 10
            WHEN ws.scaling_strength IN ('B', 'C') THEN 5
            ELSE 0
         END
         - (w.required_intelligence + w.required_faith) * 0.5
        ) AS pure_str_score,

        -- Pure DEX score
        (CASE 
            WHEN ws.scaling_dexterity IN ('S', 'A') THEN 10
            WHEN ws.scaling_dexterity IN ('B', 'C') THEN 5
            ELSE 0
         END
         - (w.required_intelligence + w.required_faith) * 0.5
        ) AS pure_dex_score,

        -- Pure INT score
        (CASE 
            WHEN ws.scaling_intelligence IN ('S', 'A') THEN 10
            WHEN ws.scaling_intelligence IN ('B', 'C') THEN 5
            ELSE 0
        END) AS pure_int_score,

        -- Pure FAI score
        (CASE 
            WHEN ws.scaling_faith IN ('S', 'A') THEN 10
            WHEN ws.scaling_faith IN ('B', 'C') THEN 5
            ELSE 0
        END) AS pure_fai_score

    FROM eldenringcatalog.gold.fact_weapon_stats ws
    INNER JOIN eldenringcatalog.gold.dim_weapons w
        ON ws.weapon_name = w.weapon_name
    WHERE ws.upgrade_level LIKE '%Standard%'
    AND upgrade_level NOT LIKE '%+%'
)

-- Top 10 Strength Weapons
-- SELECT weapon_name, category, scaling_strength, pure_str_score
-- FROM df_recommendations
-- ORDER BY pure_str_score DESC
-- LIMIT 10;

-- -- Top 10 Dexterity Weapons
-- SELECT weapon_name, category, scaling_dexterity, pure_dex_score
-- FROM df_recommendations
-- ORDER BY pure_dex_score DESC
-- LIMIT 10;

-- -- Top 10 Intelligence Weapons
-- SELECT weapon_name, category, scaling_intelligence, pure_int_score
-- FROM df_recommendations
-- ORDER BY pure_int_score DESC
-- LIMIT 10;

-- -- Top 10 Faith Weapons
SELECT weapon_name, category, scaling_faith, pure_fai_score
FROM df_recommendations
ORDER BY pure_fai_score DESC
LIMIT 10;
```

These weapons are ideal for players focusing on raw Strength damage, offering high base effectiveness without any upgrades, making them strong early-game options.**

**In Dexterity, the top base weapons reach a moderate pure Dexterity score of 5, mostly from C scaling. Weapons like Scorpion's Stinger, Dagger, Main-gauche, and Parrying Dagger (all Daggers), as well as Curseblade's Cirque (Backhand Blades) and Rakshasa's Great Katana (Great Katanas), provide good starting options for Dexterity builds. While their scaling is not as high as top Strength weapons, they remain solid choices before upgrades.**

**For Intelligence-focused builds, base weapons achieve a pure Intelligence score of 5, using B or C scaling. Examples include Demi-Human Queen's Staff, Astrologer's Staff, Academy Glintstone Staff, Azur's Glintstone Staff, Glintstone Staff, and Digger's Staff, along with Dark Moon Greatsword and Moonveil. These weapons serve as effective early-game tools for spellcasters, giving moderate damage and scaling with Intelligence without needing upgrades.**

**Regarding Faith builds, the top base weapons also reach a pure Faith score of 5 through B or C scaling.
